```python
# My Code
# Write your solution to exercise 1 here

flag = True
num_list = []
r_num = []  # in case there are more than 1 most repeated

while flag:
 
    num = int(input('Type in a number: '))

    if num != 0:
        num_list.append(num)
    else:
        smallest = min(num_list)
        biggest = max(num_list)
        length = len(num_list)
        sumAll = sum(num_list)


        non_repeated_set = set(num_list) # return a set class: i.e. {1,2,3}
        # print(f'{non_repeated_set}, non_repeated_list_type = {type(non_repeated_set)}')
        
        max_t = 0
        
        for item in non_repeated_set:
            r_time = num_list.count(item)
            if r_time >= max_t:
                max_t = r_time
                r_num.append(item)

        flag = False


print(f'smallest number: {smallest} \nbiggest number: {biggest} \nNumber of numbers: {length} \nSum: {sumAll} ')
if len(r_num) > 1:
    print(f'Most repeated: {','.join(map(str,r_num))}')  # more than 1 most repeated
else:
    print(f'Most repeated: {r_num[0]}')


```

```python
# Write your solution to exercise 2 here

import string

def caesar_encrypt(text: str, shift_value: int):
    book = string.ascii_lowercase
    book_len = len(book)
    encrypted_word = ""

    for l in text:
        l_pos = book.find(l)
        new_pos = l_pos + shift_value

        if new_pos >= book_len:
            new_pos -= book_len
         
        new_l = book[new_pos]
        encrypted_word += new_l
        # print(f'letter = {l}, pos={l_pos}, new_pos={new_pos}, new_l={new_l}')   # test
        
    
    return (encrypted_word)


def caesar_decrypt(text: str, shift_value: int):
    book = string.ascii_lowercase
    book_len = len(book)
    decrypted_word = ""

    for l in text:
        l_pos = book.find(l)
        old_pos = l_pos - shift_value

        if old_pos >= book_len:
            old_pos -= book_len
         
        old_l = book[old_pos]
        decrypted_word += old_l
        # print(f'letter = {l}, pos={l_pos}, old_pos={old_pos}, old_l={old_l}')   # test
        
    
    return (decrypted_word)





if __name__ == "__main__":
    # # encrypted
    # words_to_encrypt = [
    # 'xyz', 
    # 'abc',
    # "one",
    # "of",
    # "the",
    # "foods",
    # "i",
    # "like",
    # "is",
    # "pizza"
    # ]

    # for word in words_to_encrypt:
    #     encrypted = caesar_encrypt(word, 3)
    #     print(encrypted)

    # decryped
    secret_message = [
    "ersxliv", 
    "sri", 
    "mw", 
    "qegevsrm", 
    "gewwivspi", 
    "figeywi", 
    "mx", 
    "mw", 
    "gliet", 
    ]

    for word in secret_message:
        decrypted = caesar_decrypt(word, 4)
        print(decrypted)

    # more tests: You can also make sure that your program is working properly with these function calls. The program should only print True in these tests.
    ecrypt_and_decrypt = [
    "message",
    "encrypted",
    "and",
    "decrypted"
    ]
    # Becaues the encryption is decrypted with the same shift value as it was encrypted, all the prints are True
    for word in ecrypt_and_decrypt:
        encrypted = caesar_encrypt(word, 15)
        decrypted = caesar_decrypt(encrypted, 15)
        print(word == decrypted)

    # Test the encryption with 26 as the shift value
    # The words should not change because the shift value is the same as the alphabet length

    for word in ecrypt_and_decrypt:
        encrypted = caesar_encrypt(word, 26)
        decrypted = caesar_decrypt(word, 26)
        print(word == encrypted == decrypted)
```
