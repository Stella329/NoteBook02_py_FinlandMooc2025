## part03_33 Chessboard

Please write a function named chessboard, which prints out a chessboard made out of ones and zeroes. The function takes an integer argument, which specifies the length of the side of the board. See the examples below for details:

chessboard(3)
print()
chessboard(6)
Sample output

101

010

101

---

101010

010101

101010

010101

101010

010101

```python
# Write your solution here
def chessboard(step):
    flag = 1
    while flag <= step:
        if flag % 2 !=0:
            print('10' * (step//2) + '1'*(step%2))
        else:
            print("01" * (step//2) + '0'*(step%2))
        flag +=1

# Testing the function
if __name__ == "__main__":
    chessboard(5)
```

```python
# Solution for part03-33_chessboard

def chessboard(size):
    i = 0
    while i < size:
        if i % 2 == 0:
            row = "10"*size
        else:
            row = "01"*size
        # Remove extra characters at the end of the row 截取
        print(row[0:size])
        i += 1
```
