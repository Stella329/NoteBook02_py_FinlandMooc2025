## part04-07_shape
Please write a function named shape, which takes four arguments. The first two parameters specify a triangle, as above, and the character used to draw it. The first parameter also specifies the width of a rectangle, while the third parameter specifies its height. The fourth parameter specifies the filler character of the rectangle. The function prints first the triangle, and then the rectangle below it.

The function should call the function line from the exercise above for the actual printing out. Copy your solution to that exercise above the code for this exercise. Please don't change anything in the line function.

Some examples:

shape(5, "X", 3, "*")
print()
shape(2, "o", 4, "+")
print()
shape(3, ".", 0, ",")




```python
# standard ans
def line(size, character):  # line()：前序练习创立，不让修改
    if character == "":
        character = "*"
    print(character[0] * size)
 
def shape(y_height, y_character, a_height, a_character):
    i = 1
    while i <= y_height:
        line(i, y_character)
        i += 1
    i = a_height
    while i > 0:
        line(y_height, a_character)
        i -= 1
```

## part04-08_spruce

```python
# My Solution 2024版

def spruce(line_num):
    print("a spruce!")

    symbol = "*"
    line = int(line_num)

    for item in range(0, line):
        star_num = item*2+1 # 行数 * 2 + 1 = 星数
        space = line - (item+1)
        print(" "*space + symbol*star_num + " "*space)
        star_num += 2

    # tree trunck == 第一行
    space = line - 1
    print(" "*space + symbol + " "*space)


# You can test your function by calling it within the following block
if __name__ == "__main__":
    spruce(5)
```

```python
# My solution 2025版

def spruce(line):
    end = 1 + 2 * line # 总星数
    star = 1
    step = 0

    print('a spruce!')
    
    # tree body
    while step < line:
        space = (end - star -1 ) // 2  # 星数 = 行数 * 2 + 1
        print(space * ' '+ star * '*' + space * ' ')
        step += 1
        star += 2
    
    # trunk
    star = 1
    space = (end - star -1 ) //2
    print(space * ' '+ star * '*' + space * ' ')

# You can test your function by calling it within the following block
if __name__ == "__main__":
    spruce(5)
```

```python
# Solution for part04-08_spruce
def spruce(height):
    print("a spruce!")
    i = 1
    while i <= height:
        empty = height - i
        stars = 2 * i - 1
        print(" " * empty + "*" * stars)
        i += 1
    print(" " * (height - 1) + "*")
```

## part04-11_first_second_last: Find All

Please write three functions: first_word, second_word and last_word. Each function takes a string argument.

As their names imply, the functions return either the first, the second or the last word in the sentence they receive as their string argument.

In each case you may assume the argument string contains at least two separate words, and all words are separated by exactly one space character. There will be no spaces in the beginning or at the end of the argument strings.




```python
# my solutuion: 使用 string.split() --》返回list，切割单词

# Write your solution here
def first_word(sentence):
    space = sentence.find(' ') # 返回第一个找到的space
    return sentence[:space]

def second_word(sentence):
    words = sentence.split() # list
    return words[1]

def last_word(sentence):
    words = sentence.split() # list
    return words[-1]

# You can test your function by calling it within the following block
if __name__ == "__main__":
    sentence = "once upon a time there was a programmer"
    print(first_word(sentence))
    print(second_word(sentence))
    print(last_word(sentence))
```

```python
# standard ans
def find_word(str, whatth):
    index = 0
    word = ""
    counter = 0
    while index < len(str):
    	if str[index] == " ":
    	    counter += 1
    	    if counter == whatth:
    	        break
    	    word = ""
    	else:
    	    word += str[index]
    	index += 1
    return word
 
def first_word(mjono):
    return find_word(mjono, 1)
 
def second_word(mjono):
    return find_word(mjono, 2)
 
def last_word(mjono):
    return find_word(mjono, 0)


# You can test your function by calling it within the following block
if __name__ == "__main__":
    sentence = "once upon a time there was a programmer"
    print(first_word(sentence))
    print(second_word(sentence))
    print(last_word(sentence))
```

## part04-30_shortest_in_list

Please write a function named shortest, which takes a list of strings as its argument. The function returns whichever of the strings is the shortest. If more than one are equally short, the function can return any of the shortest strings (there will be no such situation in the tests). You may assume there will be no empty strings in the list.

my_list = ["first", "second", "fourth", "eleventh"]

result = shortest(my_list)

print(result)

---

my_list = ["adele", "mark", "dorothy", "tim", "hedy", "richard"]


result = shortest(my_list)

print(result)

---

Sample output

first

tim

```python
# my solution: list.index(a)

def shortest(my_list):

    list_word = []
    list_length = []

    for item in my_list:
        list_word.append(item)
        list_length.append(len(item))

    min_length = min(list_length)
    index = list_length.index(min_length) # find the position 返回第一个

    return list_word[index]


# my_list = ["adele", "mark", "dorothy", "tim", "hedy", "richard"]
# result = shortest(my_list)
# print(result)
```

```python
# standard solution

def shortest(names: list):
    result = ""
 
    for nimi in names:
        if result == "" or len(nimi) < len(result):
            result = nimi
 
    return result
```

## part04-31_all_longest_in_list

Please write a function named all_the_longest, which takes a list of strings as its argument. The function should return a new list containing the longest string in the original list. If more than one are equally long, the function should return all of the longest strings.

The order of the strings in the returned list should be the same as in the original.

my_list = ["adele", "mark", "dorothy", "tim", "hedy", "richard"]

result = all_the_longest(my_list)

print(result) # ['dorothy', 'richard']

```python
# standard ans

def all_the_longest(names: list):
    result = []
 
    for name in names:
        if result == [] or len(name) > len(result[0]):  # result中所有元素都是相同长度，且==result[0]
            result = [name]  # 因此如果出现更长的，可全list被替换
        elif len(name) == len(result[0]):
            result.append(name)
 
    return result
```

```python
# My solution

def all_the_longest(my_list):
    list_longestWord = []
    list_length = []

    for item in my_list:
        list_length.append(len(item))

    longest = max(list_length)

    for word in my_list:
        if len(word) == longest:
            list_longestWord.append(word)
    
    return list_longestWord


# my_list = ['Alan', 'Steve', 'Seymour', 'Kim', 'Susan']

# result = all_the_longest(my_list)
# print(result)
```
