## part05-16_histogram

Please write a function named histogram, which takes a string as its argument. The function should print out a histogram representing the number of times each letter occurs in the string. Each occurrence of a letter should be represented by a star on the specific line for that letter.

For example, the function call histogram("abba") should print out

Sample output

a **

b **

```python
def histogram(word:str):
    his_dic = {}
    for string in word:
        if string not in his_dic: # check the key
            his_dic[string] = 0
        his_dic[string] += 1
    
    for key, value in his_dic.items():
        star = "*" * value
        print(f'{key} {star}')


if __name__ == "__main__":
    histogram("abba")
    # histogram("statistically")
```

## part05-19_invert_dictionary：Copy中的Scope问题
Please write a function named invert(dictionary: dict), which takes a dictionary as its argument. The dictionary should be inverted in place so that values become keys and keys become values.

An example of its use:

s = {1: "first", 2: "second", 3: "third", 4: "fourth"}

invert(s)

print(s)

Sample output

{"first": 1, "second": 2, "third": 3, "fourth": 4}

NB: the principles regarding lists covered here also hold for dictionaries passed as arguments.

If you have trouble completing this exercise, the visualisation tool might help you understand what your code is or isn't doing.

```python
# Write your solution here

def invert(dictionary: dict):
    new_dic ={}
    for item in dictionary: # --> iterate and return the key only
        value = dictionary[item]
        new_dic[value] = item
    
    dictionary.clear()         # 清空原字典
    dictionary.update(new_dic) # 更新为新内容


if __name__ == "__main__":
    s = {1: "first", 2: "second", 3: "third", 4: "fourth"}
    invert(s) # need to change the original dict
    print(s)
```

```python
# standard ans 

def invert(dictionary: dict): # AB容器互换概念，需要C容器
	copy = {}
	for key in dictionary:
		copy[key] = dictionary[key]
	for key in copy:
		del dictionary[key]
	for key in copy:
		dictionary[copy[key]] = key
```

## part05-20_numbers_spelled_out

**NB: Please don't formulate each spelled out number by hand. Figure out how you can use loops and dictionaries in your solution.**

Please write a function named dict_of_numbers(), which returns a new dictionary. The dictionary should have the numbers from 0 to 99 as its keys. The value attached to each key should be the number spelled out in words. Please have a look at the example below:

numbers = dict_of_numbers()

print(numbers[45])

print(numbers[99])

print(numbers[0])

---
Sample output

forty-five

ninety-nine

zero

```python
# Write your solution here

def dict_of_numbers():
    dic = {}
    ones = ["", "one", "two", "three", "four", "five", "six", "seven", "eight", "nine"]
    teens = ["ten", "eleven", "twelve", "thirteen", "fourteen", "fifteen",
             "sixteen", "seventeen", "eighteen", "nineteen"]
    tens = ["", "", "twenty", "thirty", "forty", "fifty",
            "sixty", "seventy", "eighty", "ninety"]
    
    dic[0] = 'zero'

    for num in range(1,10): # 1-9
        dic[num] = ones[num]
    
    for num in range(10,20): # 10-19
        dic[num] = teens[num-10]

    for step in range(20,100): # from 20
        ten_part = step // 10 # 26 -> 2
        one_part = step % 10 # 26 -> 6
        if one_part == 0:
            dic[step] = f'{tens[ten_part]}'
        else: 
            dic[step] = f'{tens[ten_part]}-{ones[one_part]}'

    return dic

if __name__ == "__main__":
    print(dict_of_numbers())
```

```python
# standard ans

def dict_of_numbers():
    # Helper dictionaries
    singles = {0:"zero", 1:"one", 2:"two", 3:"three", 4:"four", 5:"five", 6:"six", 7:"seven", 8:"eight", 9:"nine"}
    whole_tens = {2: "twenty", 3: "thirty", 4: "forty", 5: "fifty", 6: "sixty", 7: "seventy", 8: "eighty", 9: "ninety"}
 
    numbers = {}
 
    # 0 - 9
    for i in range(10):
        numbers[i] = singles[i]
 
    # 10 - 19 are special cases
    numbers[10] = "ten"
    numbers[11] = "eleven"
    numbers[12] = "twelve"
    numbers[13] = "thirteen"
    numbers[14] = "fourteen"
    numbers[15] = "fifteen"
    numbers[16] = "sixteen"
    numbers[17] = "seventeen"
    numbers[18] = "eighteen"
    numbers[19] = "nineteen"
 
    # 20 - 99
    for i in range(2, 10):
        numbers[i * 10] = whole_tens[i]
        for j in range(1, 10):
            numbers[i * 10 + j] = whole_tens[i] + "-" + singles[j]
 
    return numbers
 
if __name__ == "__main__":
    print(dict_of_numbers())
```
