## part03-23_multiplication

Please write a program which asks the user for a positive integer number. The program then prints out a list of multiplication operations until both operands reach the number given by the user. See the examples below for details:

Sample output

1 x 2 = 2
2 x 1 = 2
2 x 2 = 4

```python
# my solution
num = int(input('Please type in a number:'))

num1=1
num2=1

while num1 <=num:
    for item in range(1,num+1): #num2
        print(f"{num1} x {item} = {num1*item}")
        item +=1
    num1+=1
```

```python
# standard solution
number = int(input("Please type in a number: "))
counter1 = 1
while counter1 <= number:
    counter2 = 1
    while counter2 <= number:
        print(f"{counter1} x {counter2} = {counter1*counter2}")
        counter2 += 1
    counter1 += 1
```

## part03-25_factorial

```python
# standard solution

while True:
    number = int(input("Please type in a number: "))
    if number <= 0:
        break
 
    factorial = 1
    new = 1
    while new <= number:
        factorial *= new
        new += 1
 
    print(f"The factorial of the number {number} is {factorial}")
 
print("Thanks and bye!")
```

```python
# my solution ()

while True:
    num = int(input('Please type in a number: '))
    factorial = num

    if num >0:
        step = num
        while step > 1:
            factorial *= (step-1)
            step -= 1
        print(f"The factorial of the number {num} is {factorial}")
    else: 
        break
print('Thanks and bye!')
```

## part03-26_flip_the_pairs

Please write a program which asks the user to type in a number. The program then prints out all the positive integer values from 1 up to the number. However, the order of the numbers is changed so that each pair or numbers is flipped. That is, 2 comes before 1, 4 before 3 and so forth. See the examples below for details.

Sample output
Please type in a number: 5

2
1
4
3
5

Sample output
Please type in a number: 6

2
1
4
3
6
5

```python
# Solution for part03-26_flip_the_pairs

number = int(input("Please type in a number: "))
 
index = 1
while index+1 <= number:
    print(index+1)
    print(index)
    index += 2  # each pair handled
 
if index <= number:
    print(index)
```

## part03-27_taking_turns

Please write a program which asks the user to type in a number. The program then prints out the positive integers between 1 and the number itself, alternating between the two ends of the range as in the examples below.

Sample output
Please type in a number: 5

4

Sample output
Please type in a number: 6

1

```python
# standard solution

number = int(input("Please type in a number: "))
 
left = 1
right = number
 
while left < right:
    print(left)
    print(right)
    left += 1
    right -= 1
 
if left == right:
    print(left)
```
