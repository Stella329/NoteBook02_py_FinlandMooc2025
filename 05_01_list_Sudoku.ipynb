```python
# standard ans

def block_correct(sudoku: list, row_no: int, column_no: int):
    numbers = []
    for r in range(row_no, row_no+3):
        for s in range(column_no, column_no+3):
            number = sudoku[r][s]
            if number > 0 and number in numbers:
                return False
            numbers.append(number)
 
    return True
```

# Part5 Section2: Reference_Sudoku

What is stored in a variable is not the value per se, but **a reference to the object which is the actual value of the variable**. The object can be e.g. a number, a string or a list.

**reference**: location info
- A reference is often represented by an arrow from the variable to the actual value in memory
- So, a reference tells us where the value can be found. The function id can be used to find out the exact location the variable points to.
- The reference, or the ID of the variable, is an integer, which can be thought of as the address in computer memory where the value of the variable is stored. 
- In reality, Python strings are handled very much like lists, with references to locations in memory.

i.e. It seems Python has stored the value 1 in the memory location 4535856912. Whenever a variable is assigned the value 1, it refers to that location in computer memory.

**It is good to keep in mind that almost everything is a reference in Python, but all this is rarely relevant to everyday programming tasks.**

You may assume the matrix is a square matrix, so it will have an equal number of rows and columns.

The function should not have a return value. The matrix should be modified directly through the reference.



```python
# my code

def transpose(matrix: list):
    new_matrix = []
    # do transposing
    for lis in range(len(matrix)):
        new_matrix.append([]) #行数

    for line in matrix:
        for num in range(len(line)):
            new_matrix[num].append(line[num])
    
    # 要求直接修改原matrix
    for x in range(len(new_matrix)):
        for y in range(len(new_matrix[x])):
            matrix[x][y] = new_matrix[x][y]
    print(matrix)


if __name__ == '__main__':
    transpose([[1, 2], [1, 2]])
```
